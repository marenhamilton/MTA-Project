```python
import pandas as pd
```

```python
#Brooklyn Tech Triangle coordinate boundaries gathered from pins on Google Maps
point1 = [40.703020, -74.010923] #NW Corner
point2 = [40.685979, -73.991146] #SW Corner
point3 = [40.6806397, -73.9747435] #SE corner
point4 = [40.698351, -73.972434] #NE Corner
```

```python
#Manhattan Silicon Alley coordinate boundaries (95% of Manhattan tech companies)

point5 = (40.762261, -73.966260) #NE corner
point6 = (40.701598, -74.012487) #SE corner
point7 = (40.767958, -73.995653) #W corner
```

```python
mta_coordinates = pd.read_csv('http://web.mta.info/developers/data/nyct/subway/StationEntrances.csv')
mta_coordinates.head()
```

```python
def max_min_coordinates(points):
    '''Function to take in coordinates of boundaries of the two tech zones. 
    Returns the max and min coordinates values to establish coordinate ranges for the two tech company clusters'''
    
    latitude = []
    longitude = []

    for point in points:
       latitude.append(point[0])
       longitude.append(point[1])

    return (max(latitude), min(latitude), max(longitude), min(longitude))
```

```python
Brooklyn_coordinates = [point1, point2, point3, point4]
Manhattan_coordinates = [point5, point6, point7]
```

```python
lat_max, lat_min, long_max, long_min = max_min_coordinates(Brooklyn_coordinates) 
#Filters out stations outside of max & min lat/long coordinates 
Brooklyn_list = mta_coordinates[(mta_coordinates['Station_Latitude'] >= lat_min) & (mta_coordinates['Station_Latitude'] <= lat_max)\
            & (mta_coordinates['Station_Longitude'] >= long_min) & (mta_coordinates['Station_Longitude'] <= long_max)]
```

```python
#get unique set of subway stations within Brooklyn Tech Triangle
BK_tech_stops = set(Brooklyn_list['Station_Name'])
BK_tech_stops = list(BK_tech_stops)
BK_tech_stops
```

```python
lat_max, lat_min, long_max, long_min = max_min_coordinates(Manhattan_coordinates) 
#Filters out stations outside of max & min lat/long coordinates 
Manhattan_list = mta_coordinates[(mta_coordinates['Station_Latitude'] >= lat_min) & (mta_coordinates['Station_Latitude'] <= lat_max)\
            & (mta_coordinates['Station_Longitude'] >= long_min) & (mta_coordinates['Station_Longitude'] <= long_max)]
```

```python
#get unique set of subway stations within Manhattan's 'Silicon Alley'
Manhattan_tech_stops = set(Manhattan_list['Station_Name'])
Manhattan_tech_stops = list(Manhattan_tech_stops)
Manhattan_tech_stops
```

```python
combined_tech_stops = sorted(Manhattan_tech_stops + BK_tech_stops)
combined_tech_stops
```
